```python
plt.boxplot(df1['Salary'])
plt.title('Box plot of Salary')
plt.show()
```

```python
Q3 = df1['Salary'].quantile(0.75)
Q1 = df1['Salary'].quantile(0.25)
IQR = Q3 - Q1
upper_limt = Q3+(1.5*IQR)
lower_limt = Q1 -(1.5*IQR)

outliner_salary = df1[(df1['Salary'] > upper_limt) |(df1['Salary'] < lower_limt)]
outliner_salary
```

```python
x = np.arange(8)
y = x-1
```

```python
z= x**2
e=y**2
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z, marker = "*")
plt.show()
```

```python
plt.plot(x,y)
plt.plot(x,z)
plt.plot(y,e)
plt.plot(x,e)
plt.plot(y,z)
plt.show()
```

```python
plt.plot(x,y, label = 'x&y')
plt.plot(x,z, label = 'x&z')
plt.plot(x,e, label = 'x&e')

plt.legend()
plt.show()
```

```python
plt.plot(y,x, label = 'y&x', marker = "*")
plt.plot(y,z, label = 'y&z', marker = "+")
plt.plot(y,e, label = 'y&e')

plt.legend()
plt.grid(True)
plt.show()
```

```python
#pie chart #bar chart
education=['B.Tech', 'M.Tech', 'B.Sc', 'M.Sc']
no_of_students_srk=[50,25,15,10] 
no_of_students_x=[60,30,5,5]
```

```python
plt.bar(education, no_of_students_srk)
plt.ylabel("no_of_students_srk")
plt.title("SRK Students")
plt.show()
```

```python
plt.pie(no_of_students_srk, labels = education, autopct ="%0.1f%%")
plt.title("% of SRK Students")

plt.show()
```

```python
plt.barh(education, no_of_students_srk)
plt.ylabel("no_of_students_srk")
plt.title("SRK Students")
plt.grid(True)
plt.show()
```
